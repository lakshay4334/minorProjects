```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## generate sin & cos wave

```python
x = np.linspace(0,3*np.pi,100)
```

```python
sin_y = np.sin(x)
cos_y = np.cos(x)
```

```python
plt.plot(x,sin_y, label = 'sin')
plt.plot(x,cos_y, label = 'cos')
plt.legend(loc = 'upper right')
plt.title('Sine Wave')
plt.show()
```

```python
plt.plot(x,sin_y, '--')
plt.plot(x,cos_y, "*" )
plt.legend(loc = 'upper right')
plt.title('Sine Wave')
plt.show()
```

## setting different styles on graph

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')
plt.plot(x,sin_y, label = 'sin')
plt.plot(x,cos_y, label = 'cos')
plt.legend(loc = 'upper right')
plt.title('Sine & Cos Wave')
plt.show()
```

## increasing the size of plot

```python
plt.figure(figsize=(18,5))
plt.style.use('seaborn-v0_8')
plt.plot(x,sin_y, label = 'sin')
plt.plot(x,cos_y, label = 'cos')
plt.legend(loc = 'upper right')
plt.title('Sine & Cos Wave')
plt.show()
```

## saving plots locally

```python
plt.figure(figsize=(18,5))
fig = plt.figure()
plt.plot(x,sin_y, label = 'sin')
plt.plot(x,cos_y, label = 'cos')
plt.legend(loc = 'upper right')
plt.title('Sine & Cos Wave')
plt.show()


fig.savefig("sine and cosine.png")
```

# Pie Charts

```python
cars = ['Audi','BMW','Ford','Tesla','Jaguar','Mercedes']
data = [450,385,35,10,200,500]

fig = plt.figure(figsize = (7,7) )
plt.pie(data, labels = cars)
plt.show()
```

```python
cars = ['Audi','BMW','Ford','Tesla','Jaguar','Mercedes']
data = [450,385,35,10,200,500]
explode = [0,0,0,0.1,0,0]

fig = plt.figure(figsize = (7,7) )
plt.pie(data, labels = cars, explode = explode)
plt.show()
```

## adding shadows

```python
cars = ['Audi','BMW','Ford','Tesla','Jaguar','Mercedes']
data = [450,385,35,10,200,500]
explode = [0,0,0,0.1,0,0]

fig = plt.figure(figsize = (7,7) )
plt.pie(data, labels = cars, explode = explode,shadow = True)
plt.show()
```

## adding percentage

```python
cars = ['Audi','BMW','Ford','Tesla','Jaguar','Mercedes']
data = [450,385,35,10,200,500]
explode = [0,0,0,0.1,0,0]

fig = plt.figure(figsize = (7,7) )
plt.pie(data, labels = cars, explode = explode,shadow = True, autopct = '%1.1f%%')
plt.show()
```

## adding wedget

```python
cars = ['Audi','BMW','Ford','Tesla','Jaguar','Mercedes']
data = [450,385,35,10,200,500]
explode = [0,0,0,0.1,0,0]

fig = plt.figure(figsize = (7,7) )
plt.pie(data, labels = cars, explode = explode,shadow = True, autopct = '%1.1f%%',wedgeprops = {'edgecolor':'Black','linewidth': 2,'antialiased':True})
plt.show()
```

## multiple subplots

```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('sine wave')

y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('cos wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('tan wave')

y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Dist')
plt.tight_layout()
plt.show()
```

```python
plt.subplots_adjust(hspace = 0.4,wspace = 0.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize = 15, ha = 'center')
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)


fig , (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')

plt.tight_layout()

plt.show()
```

# 3D plots
