```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
res = requests.get("https://en.wikipedia.org/wiki/Mahatma_Gandhi")
res.text
```

```python
soup = BeautifulSoup(res.text,"html.parser")
```

```python
heading = soup.find("h1").text
heading
```

```python
corpus = ""

for p in soup.find_all("p"):
    corpus += p.text
    corpus += "\n"

print(corpus)
```

```python
for i in range(0,363):
    corpus = corpus.replace("["+str(i)+"]","")
```

```python
print(corpus)
```

```python
title = input("Enter your input:")
title_str = title.replace(" ","+")

res = requests.get("https://www.google.com/search?q="+title_str+"+wikipedia")
soup = BeautifulSoup(res.text,"html.parser")

for sp in soup.find_all("div"):
    try:
        link = sp.find("a").get("href")
        if "https://en.wikipedia.org/wiki" in link:
            link = link[link.index("https"):link.index("&")]
            print(link)
            break
    except:
        pass
```
