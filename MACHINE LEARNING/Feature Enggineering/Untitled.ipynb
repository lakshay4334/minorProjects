```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("BostonHousing.csv")
data.head()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
x
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
print(xtrain.shape,xtest.shape)
print(ytrain.shape,ytest.shape)
```

## random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(100)
rf = rf.fit(xtrain,ytrain)

predictions = rf.predict(xtest)
```

```python
predictions.shape
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(ytest,predictions)
```

K Fold

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5,shuffle=False)
```

```python
for train_index,test_index in kfold.split(x):
    print("Train - \n", train_index, "\nTest - \n", test_index)
    print("=======\n")
```

```python
rf1= RandomForestRegressor()

n_est = [100, 200, 500]

for n_estimator in n_est:       
    error = []
    for train_index, test_index in kfold.split(x):
        
        xtrain, xtest = x[train_index], x[test_index]
        ytrain, ytest = y[train_index], y[test_index]
    
        rf1 = RandomForestRegressor(n_estimator)

        rf1 = rf1.fit(xtrain, ytrain)

        prediction = rf1.predict(xtest)
        
        error.append(mean_squared_error(ytest, prediction))
    print("Results for n_estimator: ", n_estimator)
    print(error)
    print("Average Error = ", sum(error)/len(error))
```
